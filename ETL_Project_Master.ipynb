```python
#Import dependencies

import pandas as pd
from sqlalchemy import create_engine
```

```python
#Read the individual csv files of Happiness data set and add a column for Year
#Read the csv file for suicide dataset

file_2015_path = "world-happiness/2015.csv"
happiness_2015 = pd.read_csv(file_2015_path)
happiness_2015["Year"] = "2015"

file_2016_path = "world-happiness/2016.csv"
happiness_2016 = pd.read_csv(file_2016_path)
happiness_2016["Year"] = "2016"

file_2017_path = "world-happiness/2017.csv"
happiness_2017 = pd.read_csv(file_2017_path)
happiness_2017["Year"] = "2017"

file_path_18 = "world-happiness/2018.csv"
happiness_2018 = pd.read_csv(file_path_18)
happiness_2018['Year'] = 2018

file_path_19 = "world-happiness/2019.csv"
happiness_2019 = pd.read_csv(file_path_19)
happiness_2019['Year'] = 2019

suicide_path = "who_suicide_statistics.csv/who_suicide_statistics.csv"
suicide_df = pd.read_csv(suicide_path)
```

```python
#Renaming of Column headers in the dataframe for Happiness dataset

happiness_2015_rename = happiness_2015.rename(columns={'Country': 'Country', 'Region': 'Region', 'Happiness Rank': 'Happiness_Rank', 'Happiness Score': 'Happiness_Score',
       'Standard Error': "Standard_Error", 'Economy (GDP per Capita)': 'GDP_per_capita', 'Family': 'Family',
       'Health (Life Expectancy)': 'Healthy_life_expectancy', 'Freedom': 'Freedom', 'Trust (Government Corruption)': 'Perceptions_of_corruption',
       'Generosity': 'Generosity', 'Dystopia Residual': 'Dystopia_Residual', 'Year': 'Year'})


happiness_2016_rename = happiness_2016.rename(columns={'Country': 'Country', 'Region': 'Region', 'Happiness Rank': 'Happiness_Rank', 'Happiness Score': 'Happiness_Score',
       'Lower Confidence Interval': 'Lower_Confidence_Interval', 'Upper Confidence Interval': 'Upper_Confidence_Interval',
       'Economy (GDP per Capita)': 'GDP_per_capita', 'Family': 'Family', 'Health (Life Expectancy)': 'Healthy_life_expectancy',
       'Freedom': 'Freedom' , 'Trust (Government Corruption)': 'Perceptions_of_corruption', 'Generosity': 'Generosity',
       'Dystopia Residual': 'Dystopia_Residual', 'Year': 'Year'})

happiness_2017_rename = happiness_2017.rename(columns={'Country': 'Country', 'Happiness.Rank': 'Happiness_Rank', 'Happiness.Score': 'Happiness_Score', 'Whisker.high': 'Whisker_high',
       'Whisker.low' : 'Whisker_low', 'Economy..GDP.per.Capita.': 'GDP_per_capita', 'Family': 'Family',
       'Health..Life.Expectancy.': 'Healthy_life_expectancy', 'Freedom': 'Freedom' , 'Generosity': 'Generosity',
       'Trust..Government.Corruption.': 'Perceptions_of_corruption', 'Dystopia.Residual': 'Dystopia_Residual', 'Year': 'Year'})




happiness_2018_rename = happiness_2018.rename(columns={'Country or region':'Country','Freedom to make life choices':'Freedom',
                                               'GDP per capita':'GDP_per_capita','Generosity':'Generosity',
                                               'Healthy life expectancy':'Healthy_life_expectancy','Overall rank':'Happiness_Rank',
                                                'Perceptions of corruption':'Perceptions_of_corruption',
                                                'Score':'Happiness_Score','Social support':'Social_support'})



happiness_2019_rename = happiness_2019.rename(columns={'Country or region':'Country','Freedom to make life choices':'Freedom',
                                               'GDP per capita':'GDP_per_capita','Generosity':'Generosity',
                                               'Healthy life expectancy':'Healthy_life_expectancy','Overall rank':'Happiness_Rank',
                                                'Perceptions of corruption':'Perceptions_of_corruption',
                                                'Score':'Happiness_Score','Social support':'Social_support'})

```

```python
#Concatenating of dataframes in to a single dataframe for the happiness dataset

happiness_all = pd.concat([happiness_2015_rename,happiness_2016_rename,happiness_2017_rename,happiness_2018_rename,happiness_2019_rename])

happiness_all
```

```python
#Renaming of columns in the Suicide dataset

suicide_df_rename = suicide_df.rename(columns={'country': 'Country', 'year': 'Year', 'sex': 'Sex', 
                                               'age': 'Age', 'suicides_no': 'number_of_suicides', 'population': 'population'})


suicide_df_rename
```

```python
#Filtering the suicide dataset to the columns needed
#Aggregating the suicide count for each country and year
#Created a copy(suicide_df_count) of original dataframe(suicide_df_rename) to work with



suicide_df_count = suicide_df_rename[['Country','Year','Age','number_of_suicides','population']]

suicide_groupby = suicide_df_count.groupby(['Country','Year']).agg({'number_of_suicides':'count',
                                                                        'population':'first'}).reset_index()
suicide_groupby
```

```python
#Replacing NaN values in the happiness and suicide dataframes to match not null constraint in database

happiness_all.fillna(0,inplace=True)

suicide_groupby.fillna(0,inplace=True)
```

```python
#Establishing connection to a postgres database (etl_project)

engine = create_engine('postgresql://username:password@localhost:5432/etl_project')
connection = engine.connect()
```

```python
#Check to see if the table names match
engine.table_names()
```

```python
#Load happiness data in to happiness table

happiness_all.to_sql(name='happiness', con=engine, if_exists='append', index=False)
```

```python
#Load suicide data in to suicide table

suicide_groupby.to_sql(name='suicides', con=engine, if_exists='append', index=False)
```
